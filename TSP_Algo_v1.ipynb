## 計算distance function

```python
def distance(x1,y1,x2,y2):
    d = math.sqrt((x1-x2)**2+(y1-y2)**2)
    return d
```

## 產生array function

```python
def genArray(_array_amount, _array_size, _array_maxValue):
    gen_array = []
    for i in range(_array_amount):
        array_size = (_array_size, 2)
        # 產生一組有_array_size個座標的array, 在(0-_array_maxValue)內
        array_org = np.random.randint(0, _array_maxValue, size=array_size)
        gen_array.append(array_org)
    return gen_array
```

### 獲得平均離其他點最遠的座標index

```python
# array_far = [[0,0],[3,0],[3,4],[100,100]]
```

```python
def getFarCorIndex(_array):
    array_far = _array
    far_start = 0
    total_far_d = 0
    for i in range(len(array_far)): 
        total_d = 0
        for j in range(len(array_far)):
            d = distance(array_far[i][0],array_far[i][1],array_far[j][0],array_far[j][1])
            total_d = total_d + d
#         print("第",i ,"個:", total_d)
        if (total_d > total_far_d):
            total_far_d = total_d
            far_start = i
    print("最長d：",total_far_d,"第幾個：", far_start, "座標：", array_far[far_start])
    return far_start
```

### 獲得平均離其他點最近的座標index

```python
def getNearCorIndex(_array):
    array_near = _array
    near_start = 0
    total_near_d = 0
    for i in range(len(array_near)): 
        total_d = 0
        for j in range(len(array_near)):
            d = distance(array_near[i][0],array_near[i][1],array_near[j][0],array_near[j][1])
            total_d = total_d + d
#         print("第",i ,"個:", total_d)
        if (total_near_d == 0 or total_d < total_near_d):
            total_near_d = total_d
            near_start = i
    print("最短d：",total_near_d,"第幾個：", near_start, "座標：", array_near[near_start])
    return near_start
```

## greedy algorithm function

```python
def greedy_TSP(_start_point_index, _array_org):
    array_greedy = _array_org
    x_set = array_greedy[_start_point_index][0]
    y_set = array_greedy[_start_point_index][1]
    # print(x_set)
    # print(y_set)
    TSP_List_x = []
    TSP_List_y = []
    TSP_List_x.append(x_set)
    TSP_List_y.append(y_set)
    total_d = 0
    smallest_d = 0
    smallest_cor = 0
    delete_index = _start_point_index
    
    while 1 :
        #刪除後繼續跑
        array_greedy = np.delete(array_greedy, delete_index, axis=0)
#         print("new",array_greedy)
        if len(array_greedy)==0:
            #加入第一個點以便畫圖
            TSP_List_x.append(TSP_List_x[0])
            TSP_List_y.append(TSP_List_y[0])
            total_d = total_d + distance(TSP_List_x[len(TSP_List_x)-1],TSP_List_y[len(TSP_List_y)-1],TSP_List_x[len(TSP_List_x)-2],TSP_List_y[len(TSP_List_y)-2])
            break

        for i in range(len(array_greedy)):
        #     print("x:",i[0]) 
        #     print("y:",i[1])
            d=round(distance(x_set,y_set,array_greedy[i][0],array_greedy[i][1]),2)
            if d < smallest_d or smallest_d == 0:
                smallest_d = d 
                smallest_cor = array_greedy[i]
                delete_index = i
                #print("smallest distance",smallest_d)
                #print("smallest coordinate",smallest_cor)
        #print("smallest distance",smallest_d)
        total_d = total_d + smallest_d
        TSP_List_x.append(smallest_cor[0])
        TSP_List_y.append(smallest_cor[1])
    
        # 換成下一個
        x_set = smallest_cor[0]
        y_set = smallest_cor[1]
        smallest_d = 1000000
    
#     x, y = zip(*_array_org)
#     plt.scatter(x, y)
#     plt.show()
#     plt.plot(TSP_List_x,TSP_List_y)
#     plt.show()
    print("total distance:",total_d)
    return total_d
```

### 產生Array

```python
arrayAmount = 10000  #array的組數
arraySize = 200  #一組array內的座標數量
arrayMaxValue = 100  # array內的座標最大值(0-arrayMaxValue)
array = genArray(arrayAmount, arraySize, arrayMaxValue)
# print(array)
```

### 隨機起始點

```python
random_total = 0
for i in range(arrayAmount):
    ran_start = np.random.randint(arraySize)
    random_total = random_total + greedy_TSP(ran_start, array[i])
    print("random start cor:", ran_start)
random_avg = random_total / arrayAmount
```

```python
print("random total:", random_total)
print("random average:", random_avg)
```

### 從平均距離其他點最遠的點當作起始點

```python
far_total = 0
for i in range(arrayAmount):
    far_start = getFarCorIndex(array[i])
    far_total = far_total + greedy_TSP(far_start, array[i])
far_avg = far_total / arrayAmount 
```

```python
print("furthest total:", far_total)
print("furthest average:", far_avg)
```

### 從平均距離其他點最近的點當作起始點

```python
near_total = 0
for i in range(arrayAmount):
    near_start = getNearCorIndex(array[i])
    near_total = near_total + greedy_TSP(near_start, array[i])
near_avg = near_total / arrayAmount 
```

```python
print("nearest total:", near_total)
print("nearest average:", near_avg)
```
